## Read in json from BIDS

```python
# Take input from specified path
json_from_bids = '/content/drive/MyDrive/nenah-bids-slicetiming/code/NENAH-BIDS/test_data/sub-NENAH012_acq-gre_dir-AP_run-1_phasediff.json'

# Take input from argument
# json_from_bids = str(sys.argv[1])

with open(json_from_bids) as jsonFile:
    jsonObject = json.load(jsonFile)
    jsonFile.close()
```

## Next, add IntendedFor
See [here](https://bids-specification.readthedocs.io/en/stable/04-modality-specific-files/01-magnetic-resonance-imaging-data.html#using-intendedfor-metadata)

```python
boldfile_list = []
test_list = ['a', 'sub-NENAH012_acq-gre_dir-AP_run-1_phasediff.json', 'sub-NENAH012_task-rest_dir-AP_run-1_bold.nii.gz', 'sub-NENAH012_task-rest_dir-AP_run-2_bold.nii.gz']

subj = str(test_list[1]).split('_')[0]

for boldfile in test_list[2:]:
  boldfile_list.append(subj + '/func/' + boldfile)
print(boldfile_list)
```

```python
# jsonObject['IntendedFor'] = ["func/sub-NENAH012_task-rest_dir-AP_run-1_bold.nii.gz"]

# Add list of files
jsonObject['IntendedFor'] = boldfile_list
print(jsonObject['IntendedFor'])
```
